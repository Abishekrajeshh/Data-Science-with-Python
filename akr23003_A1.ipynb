# üîµ Efficiency by vehicle

```python
df = pd.read_html('https://en.wikipedia.org/wiki/Plug-in_electric_vehicles_in_the_United_States')[4]
df
```

I'll help you get started - real-world data is messy. See the header? See the last row? These will cause problems for you. The multi-index is ANNOYING for selecting the column you want, and the last row is a text cell that repeats for each column. This is a little advanced for new coders, but you have to have good instincts.

* https://towardsdatascience.com/how-to-flatten-multiindex-columns-and-rows-in-pandas-f5406c50e569
* https://stackoverflow.com/questions/26921651/how-to-delete-the-last-row-of-data-of-a-pandas-dataframe

```python
# drop the last row
df = df[:-1]

# collapse the multi-index
df.columns = df.columns.get_level_values(0)

# check your work
df.head()
```

See how the data is dirty? You need to clean it up! Get rid of special characters like dollar signs, periods, commas and anything else that is annoying! I leave this to you to figure out.

```python
df.columns
```

```python
df.info()
```

This one was hard for me too - there was a special character HIDING in this column name!

```python
# look at how nasty this column name secretly is!
df.columns[7]
```

```python
# rename the column
df.rename({df.columns[7]  :'Annual fuel cost'}, axis='columns', inplace=True)
# check your work
df.head()

```

```python
# remove special characters
df['Annual fuel cost'] = df['Annual fuel cost'].str.replace('$', '')
df['Annual fuel cost'] = df['Annual fuel cost'].str.replace('-', '')
df['Annual fuel cost'] = df['Annual fuel cost'].str.replace(',', '')

df['Cost to drive 25 miles'] = df['Cost to drive 25 miles'].str.replace('$', '')
df['Cost to drive 25 miles'] = df['Cost to drive 25 miles'].str.replace('-', '')
df['Cost to drive 25 miles'] = df['Cost to drive 25 miles'].str.replace(',', '')

df.head()
```


**Data Cleaning Part**

```python
df['Vehicle'] = df['Vehicle'].str.split('[').str[0] # First column is cleaned
df['Operating mode (AER)'] = df['Operating mode (AER)'].str.split('(').str[0].str.strip() #Third column is cleaned
df.columns.values[3] = 'Combined-FuelEconomy' # Renaming the third column
df['Combined-FuelEconomy'] = df['Combined-FuelEconomy'].str.split('m').str[0].str.strip() # to have only numeric vales, I have removed all the charceters after M in the'mpg'
df.columns.values[4] = 'City-FuelEconomy'#Renaming the fourth column
df['City-FuelEconomy'] = df['City-FuelEconomy'].str.split('m').str[0].str.strip()
df.columns.values[5] = 'Highway-FuelEconomy'#Renaming the fifth column
df['Highway-FuelEconomy'] = df['Highway-FuelEconomy'].str.split('m').str[0].str.strip()
df
```

With special characters, the column is hiding as a string - so you can't do math on it! You need to convert in to numeric. And once it's numeric, you can make cool plots and compute summary stats.

```python
# convert column to numeric
df['Cost to drive 25 miles'] = pd.to_numeric(df['Cost to drive 25 miles'])
```

```python
# convert column to numeric
df['Model year'] = pd.to_numeric(df['Model year'], errors='coerce')
```

```python
# convert column to numeric
df['Annual fuel cost'] = pd.to_numeric(df['Annual fuel cost'])
df['Annual fuel cost'].hist()
```

## üî¥ Interesting Output 4

```python
#Heatmap - Correlation matrix for numerical columns
correlation_matrix = df[['Cost to drive 25 miles', 'Annual fuel cost']].corr()
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

The heatmap visually represents the correlation between the selected numeric columns. Positive values indicate a positive correlation, while negative values indicate a negative correlation.



A value close to 1 or -1 suggests a strong correlation, while values close to 0 suggest a weak correlation.


Understanding the correlation matrix helps identify relationships between variables, which can be valuable in data analysis and decision-making.




Features with high positive or negative correlations might influence each other, and this information can guide further exploration or modeling decisions.

## üî¥ Interesting Output 5

In this histogram displays the distribution of annual fuel costs for the vehicles in the dataset

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.histplot(df['Annual fuel cost'].dropna(), bins=20, kde=False, color='skyblue', edgecolor='black')
plt.title('Distribution of Annual Fuel Costs', fontsize=16)
plt.xlabel('Annual Fuel Cost ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The histogram provides a visual representation of the distribution of annual fuel costs.

The x-axis represents the annual fuel cost in dollars, and the y-axis represents the frequency (or count) of each cost range.


The distribution shape can reveal information about the central tendency, spread, and potential outliers in the annual fuel cost data.


Insights into common cost ranges, the presence of peaks or gaps, and the overall spread of the data can be gained from the visualization.

```python
# Now, we create a trend graph (line plot) for the average combined fuel economy over the model years
plt.figure(figsize=(15, 10))
sns.lineplot(data=df, x='Model year', y='Combined-FuelEconomy', marker='o')
plt.title('Trend of Average Combined Fuel Economy Over Model Years')
plt.xlabel('Model Year')
plt.ylabel('Average Combined Fuel Economy (mpg or mpg-e)')
plt.grid(True)
plt.tight_layout()
```
